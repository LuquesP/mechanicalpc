```python
import os 
from OCC.Extend.DataExchange import read_step_file
from OCC.Core.Tesselator import ShapeTesselator
from visualize import pcshow 
from tesselation import Tesselation
import numpy as np
import random
os.getcwd() 
```

```python
test_file = "../data/examples/Example_2.stp"
test_stp = read_step_file(test_file)  
```

```python
tess = Tesselation(filepath=test_file, format="step", outputsize=2048) 
pc  = tess.get_pointcloud() 
pc
```

```python
# Normalize PC
def normalize_pc(pointcloud): 
    norm_pointcloud = pointcloud - np.mean(pointcloud, axis=0)
    norm_pointcloud /= np.max(np.linalg.norm(norm_pointcloud, axis=1))
    return norm_pointcloud
```

```python
n_pc = normalize_pc(pc) 
pcshow(*n_pc.T) 
```
